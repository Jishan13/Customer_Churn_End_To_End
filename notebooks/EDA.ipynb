### Univariate Analysis

```python
sns.histplot(data_df["Age"], bins = 40, kde = True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#### Salary Distribution
sns.histplot(data_df["EstimatedSalary"], bins = 100, kde = True)
plt.title("Salary Estimates")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(data_df["CreditScore"], bins = 30, kde = True)
plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.ylabel("Count")
plt.show()

## Little left scewed, but most of the people are in mid range of credit score
```

```python
sns.set(style="whitegrid")
country_colors = {'France': 'blue', 'Germany': 'red', 'Spain': 'green'}
# Plotting distribution of customers across countries
plt.figure(figsize=(10, 6))
sns.countplot(x='Geography', data=data_df,palette=country_colors)
plt.title('Distribution of Customers Across Countries')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.show()

#France has the most customers
```

```python
exited_counts = data_df[data_df['Exited'] == 1].groupby('Geography')['Exited'].count()
total_counts = data_df.groupby('Geography')['CustomerId'].count()
exited_percentage = (exited_counts/total_counts) * 100
exited_percentage
```

```python
# Plotting a bar chart for churn rates by country
plt.figure(figsize=(10, 6))
sorted_exited_percentage = exited_percentage.sort_values(ascending=False)
sns.barplot(x=sorted_exited_percentage.index, y=sorted_exited_percentage,palette=country_colors)
plt.title('Churn Rates by Country')
plt.xlabel('Country')
plt.ylabel('Churn Rate (%)')
plt.show()


## Most customers are churned in Germany. So, special attention can be given there.
```

```python
gender_exited_count = data_df[data_df['Exited'] == 1].groupby('Gender')['Exited'].count()
total_count_gender_specific = data_df.groupby('Gender')["CustomerId"].count()
customer_churn_by_gender = (gender_exited_count / total_count_gender_specific) * 100
customer_churn_by_gender
country_colors = {'France': 'blue', 'Germany': 'red', 'Spain': 'green'}
# Plotting distribution of customers across countries
plt.figure(figsize=(10, 6))
sns.countplot(x='Geography', data=data_df,palette=country_colors)
plt.title('Distribution of Customers Across Countries')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.show()
## Females churn more
```

```python
churn_rate_complaints = data_df.groupby('Complain')['Exited'].mean()
churn_rate_complaints
#data_df["Complain"].value_counts()
```

```python
# Average Tenure for Churned vs. Retained Customers
avg_tenure_churned = data_df[data_df['Exited'] == 1]['Tenure'].mean()
avg_tenure_retained = data_df[data_df['Exited'] == 0]['Tenure'].mean()

# Visualizing Tenure Distribution
sns.boxplot(x='Exited', y='Tenure', data=data_df)
plt.title('Tenure Distribution by Churn Status')
plt.show()
```

```python
# Churn Rate by Number of Products
churn_rate_by_product = data_df.groupby('NumOfProducts')['Exited'].mean()

# Visualizing Product Distribution
sns.countplot(x='NumOfProducts', hue='Exited', data=data_df)
plt.title('Product Distribution by Churn Status')
plt.show()
```

```python
correlation_matrix = data_df.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
